# 准备RGB图片序列

```python
%autoreload
  
dataPath = ['/data/shared_dataset/standard_checkerboard_12x9_exposure_c15000_r1240']

key_list = []
key_list.append('standard_checkerboard_12x9_exposure_c15000_r1240')

for key in key_list:
    
    dataset_list = TianmoucDataReader(dataPath, N = 1, matchkey = key, camera_idx = 0, strict=False)

    img_list = [] # 用于保存可视化图片序列

    # 仅处理部分COP数据
    index_COP_list = [20]
    index_COP_list = range(0, len(dataset_list), 10)
    for index_COP in index_COP_list:
        sample  = dataset_list[index_COP]
        #### 处理COP数据
        rgb = sample['F0']
        rgb = torch.flip(rgb, [1]) # 镜像翻转
        img_list.append(rgb)       # 直接添加, 无需其他处理

        # 实时打印代码执行进展
        if index_COP % 200 == 0:
            print('process: '+key+'; COP_index: '+str(index_COP)+'/'+str(len(dataset_list))+'.')
            
print('done')
```

```python
%autoreload

board_size   = (12, 9) # 记录棋盘格数目
square_size  = 7.45    # 单位mm
calib_result = calibration_opencv(img_list, board_size, square_size, en_draw = 1, factor_resize = 2, en_time = 1, en_SB = 1)
```
